```python
player1 = input("Welcome to Tic Tac Toe \nPlayer 1: Would you like to be 'X' or 'O'")
if player1 == 'X':
    player2 = 'O'
else:
    player2 = 'X'
        
ready = input('Player 1 will go first\nAre you ready to play enter Yes or No?')
game = True

board=[' ',' ',' ',' ',' ',' ',' ',' ',' ']

#[0,1,2]
#[3,4,5]
#[6,7,8]
#--------------------checkEmpty-----------------
def checkEmpty(board,num):
    if board[num]==' ':
        return True
    else:
        return False
#--------------------checkIfWon-----------------
    
def checkIfWon(board):
    #rows   
    for i in range(3):
        if board[0+i] == board[3+i] == board[6+i] and board[0+i] != ' ':
            return True
        
    
     #cols
    for i in range(3):
        if board[i*3] == board[i*3+1] == board[i*3+2] and board[i*3] != ' ':
            return True
       
     #diagonal
    if board[2] == board[4] == board[6] and board[2] != ' ':
        return True
    if board[0] == board[4] == board[8] and board[0] != ' ':
        return True    
    
    
    return False


#--------------------boardPrint-----------------
def boardPrint(board):   
    for i in range(9):
        
        if i%3 == 0:
            print ('\n')
        print(board[i],end = ' |')    
        
 #--------------------Game-----------------   
boardPrint(board)
if ready=='Yes':
    while game:
        #Player1
        position = int(input('\n \n{} Choose your next position 1-9'.format(player1)))
        position -= 1
        while checkEmpty(board,position) == False:
            position = int(input('Choose your next position 1-9'))
        board[position] = player1
        if checkIfWon(board) == True:
            game =False
            boardPrint(board)
            print('\nPlayer {} Won!!!!!!!'.format(player1))
        else:    
            boardPrint(board)
        
        
        #Player2
        if game == True:
            position = int(input('\n{} Choose your next position 1-9'.format(player2)))
            position -= 1
            while checkEmpty(board,position) == False:
                position = int(input('Choose your next position 1-9'))
            board[position] = player2
            if checkIfWon(board) == True:
                game = False
                boardPrint(board)
                print('\nPlayer {} Won!!!!!!!'.format(player2))
            else:   
                boardPrint(board)
    
        if game == False:
            again = input('Do you want play again? Yes ot No')
            if again == 'Yes':
                game = True
                board=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
                


    


```
